### Factorial

n! = 1 x 2 x3 x4 x .... x n 

2! = 1 x 2 = 2

3! = 1 x2 x3 = 6

### Combination

C(n,r) = $\frac{n!}{(n-r)!r!}$

C(5, 2) = $\frac{5!}{3! * 2!} = \frac{1*~2*~3*~4*~5}{1*~2*3*~1*~2} = \frac{4*~5}{1*~2} = \frac{2*~5}{1} = 10$



```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

heights = [76, 78, 79, 81, 86]
players = ['A','B','C','D','E']
data = {x:y for x,y in zip(players, heights)}
# print(data)
print(f'The mean height of all the players is:{np.mean(heights):>5}')
print('==='*40)
print(f'The mean height of players A, B, C, D is:{np.mean((76,78,79,81)):>5}')
print('==='*40)
print(f'The mean height of player B, C, D, E is:{np.mean((78,79,81,86)):>5}')
```

```
from itertools import combinations 
def get_values(dict1: dict, search: int) -> str:
  for name, val in dict1.items():    
    if val == search:
        return name

def rSubset(arr, r): 
    return list(combinations(arr, r))

sets = rSubset(heights, 4)
data1 = {(get_values(data,x),get_values(data,y),get_values(data,z),get_values(data,l)):(x,y,z,l) for x,y,z,l in sets}
df1 = pd.DataFrame(data1.values())
df1['Samples'] = data1.keys()
cols = df1.columns.tolist()
cols = cols[-1:] + cols[:-1]
df1 = df1[cols]
df1['Sample_mean'] = df1.iloc[:,1:5].apply(np.mean,axis=1)
df1['Population_mean'] = [80]*5
display(df1)
```

## Ex1: According to a study, brain weights of Swedish men are normally distributed with a mean of 1.40 kg and a standard deviation of 0.11 kg.

### a. Determine the sampling distribution of the sample mean for samples of size 3, samples of size 12

- **For samples of size 3**:

  > $\mu_{\bar{x}} = \mu = 1.40$

  > $\sigma_{\bar{x}} = \frac{\sigma}{\sqrt{n}} = \frac{0.11}{\sqrt{3}} = 0.0635$

- **For samples of size 12**:

  > $\mu_{\bar{x}} = \mu = 1.40$

  > $\sigma_{\bar{x}} = \frac{\sigma}{\sqrt{n}} = \frac{0.11}{\sqrt{12}} = 0.0318$

### b. Determine the percentage of all samples of Swedish men that have mean brain weights within 0.1 kg of the population mean brain weight of 1.40kg for sample size of 3 and sample size of 12.

1. **Given $Z = \frac{X - \mu_{\bar{x}}}{\sigma_{\bar{x}}}$, for samples of size 3:**

$P(1.3 \leq X \leq 1.5) = P(X \leq 1.5) - P(X \leq 1.3)$

$= P(Z \leq \frac{1.5 - \mu_{\bar{x}}}{\sigma_{\bar{x}}}) - P(Z \leq \frac{1.3 - \mu_{\bar{x}}}{\sigma_{\bar{x}}}) = P(Z \leq \frac{1.5 - 1.40}{0.0635}) - P(Z \leq \frac{1.3 - 1.40}{0.0635}) = P(Z \leq 1.57) - P(Z \leq -1.57) = 0.94179 - 0.05821 = 0.88358$
  - **Intepretation: There is about 88.36% chance that the sampling error made in estimating the mean birth weight of all male babies with samples of size 3 will be at most 0.1kg**

2. **For samples of size 12:**

$P(1.39 \leq X \leq 1.41) = P(X \leq 1.41) - P(X \leq 1.39)$

$=P(Z \leq 3.14) - P(Z \leq -3.14) = 0.99916 - 0.00084 = 0.9983$

  - **Intepretation: There is about 99.83% chance that the sampling error made in estimating the mean birth weight of all male babies with samples of size 12 will be at most 0.1kg**

```
from typing import List
import numpy as np
import scipy.stats as st
np.set_printoptions(precision=2)

def get_dist(mu: float, sigma: float, n: int) -> (float, float):
  mu_bar = mu
  sigma_bar = sigma / np.sqrt(n)
  return (mu_bar, sigma_bar)

def z_score(x: float, mu: float, sigma: float) -> float:
  return (x - mu) / sigma

def z_prob(x, prob=True) -> float:
  if prob:
    return st.norm.cdf(x)
  else:
    return st.norm.ppf(x)
m = 55.4
s = 9.2
n1 = 64
n2 = 256

def prob_within(x1: float, x2: float, mu:float, sigma:float, n:int) -> float:
  mu, sigma = get_dist(mu, sigma, n)

  z1 = z_score(54.4, mu, sigma)
  z2 = z_score(56.4, mu, sigma)

  return z_prob(z2) - z_prob(z1)

prob_within(54.4, 56.4, m, s, 64)
prob_within(54.4, 56.4, m, s, 256)


```

### Ex2: The mean annual salary of public classroom teachers is \$55.4 thousand with a standard deviation of \$9.2 thousand.  

a. Find the distribution for the sample mean when the sample size is 64, 256

- **For samples of size 64**:

  > $\mu_{\bar{x}} = \mu = 55.4$

  > $\sigma_{\bar{x}} = \frac{\sigma}{\sqrt{n}} = \frac{9.2}{\sqrt{64}} = 1.15$

- **For samples of size 256**:

  > $\mu_{\bar{x}} = \mu = 55.4$

  > $\sigma_{\bar{x}} = \frac{\sigma}{\sqrt{n}} = \frac{9.2}{\sqrt{256}} = 0.575$

b. Do we need to assume that the sample mean follows a normal distribution?

c. What is the probability that the sampling error made in estimating the population mean salary will be at most $1000?

```
x4 = np.random.normal(55.4, 1.15, size=1000)
x3 = np.random.normal(55.4, 0.575, size=1000)
plt.figure(figsize=(12,10))
sns.distplot(x4, label = 'Sample size: 64')
sns.distplot(x3, label = 'Sample size: 256')
plt.bar(54.4, 1, width=0.025, color='green')
plt.bar(55.4, 1, width=0.025, color='green')
plt.bar(56.4, 1, width=0.025, color='green')
plt.legend(fontsize='large')
```

$P(54.4 \leq X \leq 56.4) = P(X \leq 56.4) - P(X \leq 54.4)$

- **For sample size of 64:**

  $P(X \leq 56.4) - P(X \leq 54.4) = 
  P(Z \leq \frac{56.4 - \mu_{\bar{x}}}{\sigma_{\bar{x}}}) - P(Z \leq \frac{54.4 - \mu_{\bar{x}}}{\sigma_{\bar{x}}}) = 
  (Z \leq \frac{56.4 - 55.4}{1.15}) - P(Z \leq \frac{54.4 - 55.4}{1.15}) = 
   P (Z \leq 0.87) - P(Z \leq -0.87) = 0.6154$

- **Similarly, for sample of size 256:**

  $P(X \leq 56.4) - P(X \leq 54.4) = 0.9179$


```
print(prob_within(54.4, 56.4, m, s, 64))
print(prob_within(54.4, 56.4, m, s, 256))
```
